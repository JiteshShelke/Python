# EXCEPTION HANDLING 

<!-- When you write a Python program, errors (also called exceptions) can happen while your 
program is running. Instead of your program crashing, exception handling helps you catch 
and manage these errors so your program can continue running smoothly. -->

```python
try:
    n=0
    res=10/n
except ZeroDivisionError:
    print('cant possible to divide by zero')
```

```python
try:
    n=10                            # code might be cause error
    res=n/0
except ZeroDivisionError:           # manage error
    print('cant divide by zero')
else:                               # run if no exception occure
    print('res is',res)
finally:                            # run always
    print('execution completed')
```

```python
# value error

try:
    a="jitesh"
    res=int(a)
except ValueError:
    print('value data type is wrong')
else:
    print(res)
```

```python
# type error

try:
    res="10"+10
except TypeError:
    print('invalid operation between data types')
else:
    print(res)
```

```python
# Index Error

try:
    ls=[10,20,30]
    res=ls[4]
except IndexError:
    print('list out of range')
```

```python
# Key Error

try:
    dictt={10:100,20:200}
    res=dictt[30]
except KeyError:
    print('key is invalid')
else:
    print(res)
```
